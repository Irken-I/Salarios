```python
import pandas as pd
import seaborn as sns

df = pd.read_csv(r'C:\Users\Insec\OneDrive\Escritorio\Todo\Salarios\Libro.csv')


df['FullIncome'] = df.Count * df.Income

countList = []
incomeList = []
fullIncomeList = []
df.head()
```

```python
binSize = 100

for i in range(0, len(df), binSize):
    binSum = df['Count'][i:i+binSize].sum()
    binSum2 = df['FullIncome'][i:i+binSize].sum()
    countList.append(round(binSum,2))
    fullIncomeList.append(round(binSum2,2))
    
```

```python
print(countList)
print(fullIncomeList)

for i in range(len(countList)):
    newValue = fullIncomeList[i]/countList[i]
    incomeList.append(round(newValue,2))
print(incomeList)
```

```python
percentageList = []
newValue2 =0

for i in range(len(countList)):
    newValue2 += countList[i]/sum(countList)*100
    percentageList.append(round(newValue2,2))

print(percentageList)

```
